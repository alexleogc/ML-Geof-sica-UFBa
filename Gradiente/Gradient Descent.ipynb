## <center> Algoritmo Gradiente (Gradient Descent) - Draft </center>

Suponha que um conjunto de observações ($y$) pode ser descrito através de uma função $\widehat{y}$:

$$ \widehat{y} = \theta_0 + \theta_1 x_1 + \theta_2 x_2 + ... +\theta_m x_m $$

matricialmente, podemos escrever y como:
$$
\begin{equation} 
\widehat{\textbf{$y$}}= \begin{bmatrix} \theta_0 & \theta_1  & \theta_2 & ... & \theta_m \end{bmatrix} \begin{bmatrix} \textbf{$u$}\\ \textbf{$x$}_1 \\  \textbf{$x$}_2 \\  \vdots \\  \textbf{$x$}_m  \end{bmatrix}
\end{equation}$$

$\textbf{$\theta$} = [\theta_0 \; \theta_1  \; \theta_2 \; ... \; \theta_m]$

$\textbf{X}^t = [\textbf{$u$},\textbf{$x$}_1,\textbf{$x$}_2,...,\textbf{$x$}_m]$

$\textbf{$u$}^t = [1 \; 1 \; ... \; 1]$, ou seja é um vetor unitário de tamanho $n$ que é o número de observações ; e

$\textbf{$x$}_m$ é um vetor de tamanho $n$.

Função para estimar y a partir de coeficientes $\textbf{$\theta$}$ e da matriz de observação $\textbf{X}$:

```python
def predict(theta,X):
    """
    Função calculada:
    X - matriz de observações
    theta - variáveis da função calculada
    """
    X = np.asarray(X) ; theta = np.asarray(theta)
    return X.dot(theta) #Calcula o produto matricial entre X e theta
```

Nestes casos, pode-se obter os valores das constantes do vetor $\textbf{$\theta$}$ através de algoritmos do tipo gradiente (neste caso, gradient descent), o qual objetiva minimizar uma função custo. Um exemplo deste tipo de função é o erro médio quadrático ($mse$) que é obtido da seguinte forma:

.

$$ mse = \frac{\sum_{i=1}^{m}(\widehat{\textbf{$y$}}-\textbf{$y$})}{m} $$

```python
def cost(theta,X,y):
    """
    Função custo que computa os desvios entre o dado observado e o calculado:
    X - Matriz das variáveis observadas (nxm : n - número de variáveis ; m - número de observações)
    y - Vetor de com as respostas esperadas (tamanho m) 
    """
    return np.sum(np.square(predict(theta,X)-y))/len(y)
```

Sabe-se que o vetor gradiente, obtido de uma função análitica ou discreta aponta na direção do valor minimo da função. Sendo assim, como queremos minimizar a função custo ($mse$), podemos escrever o vetor gradiente da seguinte forma:



```python
def gradient_descent(X,y,theta,alpha=0.01,ninter=100):
    
    try: X = np.append(X.reshape((len(X),1)),(np.ones((len(X),1))),axis=1)
    except: X = np.append(X,(np.ones((len(X),1))),axis=1)

    theta = np.asarray(theta)
    cost_history = np.zeros(ninter) ; theta_history = np.zeros((ninter,2))
    for i in range(ninter):
        theta = theta - alpha*(X.T.dot((predict(theta,X))-y))*(1/len(y))
        theta_history[i,:] = theta.T
        cost_history[i] = cost(theta,X,y)
    return theta, cost_history, theta_history
```
